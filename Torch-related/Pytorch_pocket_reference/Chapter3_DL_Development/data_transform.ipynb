Data prep taught us about pulling data form source, create dataset objects that contain info about the data itself. However data might need to be adjusted to suit training better such as normalisation, etc. These operations are called transforms.

Beauty of using pytorch for transforms is we can define a sequence of transforms and apply it when data is accessed

```python
from torchvision import transforms
from torchvision.datasets import CIFAR10

train_transforms = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=(0.49114, 0.4822, 0.4465),
        std=(0.2023, 0.1994, 0.2010))])

train_data = CIFAR10(root="./train/",
                    train=True,
                    download=True,
                    transform=train_transforms)

```

```python
print(train_data)
print(train_data.transforms)
```

```python
data, label = train_data[0]

print(type(data))

print(data.size())

print(data)
```

```python
from torchvision.datasets import CIFAR10

test_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        (0.4914, 0.4822, 0.4465),
        (0.2023, 0.1994, 0.2010))])

test_data = CIFAR10(
    root="./test/",
    train=False,
    download=True,
    transform=test_transforms)
```
